## Visualization with matplotlib

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df_city = pd.read_csv('city_data.csv')
df_city.head()
```

```
df_ride = pd.read_csv('ride_data.csv')
df_ride.head()
```

```
df = df_city.merge(df_ride.groupby(['city']).mean()['fare'], on=['city'])
df = df.merge(df_ride.groupby(['city']).count()['ride_id'], on=['city'])
df1 = df[['city', 'fare', 'driver_count', "ride_id"]]
df1.columns = ['city', 'average fare', 'number of drivers', 'number of rides']
df1 = df1.sort_values('city')
df1.head()
```

```
plt.scatter(df1['number of rides'], df1['average fare'],alpha=0.5,s=10*df1['number of rides'])
plt.xlabel('number of rides')
plt.ylabel('average fare per city')
plt.title('ride vs fare')
plt.show()
```

```
df2 = df.groupby('type').sum()
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].set_title('% of total fares')
axes[0].pie(df2['fare'], labels=df2.index, explode = (0, 0, 0.1), autopct='%1.1f%%')
axes[1].set_title('% of total rides')
axes[1].pie(df2['ride_id'], labels=df2.index, explode = (0, 0, 0.1), autopct='%1.1f%%')
```

## Dimensionality Reduction

```
df_wine = pd.read_csv('wine.data.csv')
df_wine.head()
```

```
wine_label = df_wine['Label']
df_wine = df_wine.drop('Label', axis=1)
```

```
sns.pairplot(df_wine)
```

```
cm = df_wine.corr()
sns.heatmap(cm, square=True, cmap='coolwarm')
```

```
from sklearn import preprocessing

standardScaler = preprocessing.StandardScaler()
standardScaler.fit(df_wine)
X_scaled_array = standardScaler.transform(df_wine)
normalizedData = pd.DataFrame(X_scaled_array, columns = df_wine.columns)
normalizedData
```

```
from sklearn.cluster import KMeans
import random 
  
seed = random.seed() 
kMeansClustering = KMeans(n_clusters = 3, random_state=seed)
res = kMeansClustering.fit_predict(normalizedData)
normalizedData["cluster"] = res.astype('float64')
sns_plot = sns.pairplot(normalizedData, hue = "cluster",diag_kind="hist")
```

```
from sklearn.metrics.cluster import adjusted_rand_score

adjusted_rand_score(wine_label, res)
```
