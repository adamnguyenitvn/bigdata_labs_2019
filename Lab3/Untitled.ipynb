```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('This is my Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, "r--")
plt.subplot(1,2,2)
plt.plot(y,x,'g*-');
```
